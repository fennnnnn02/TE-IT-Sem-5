```python
df.Spam[df['Spam']=='spam'] = 1
df.Spam[df['Spam']=='ham'] = 0 
```

```python
df['Text'].values
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Text'].values)
#print(vectorizer.get_feature_names_out())

X_df = pd.DataFrame(data = X.toarray(), columns = vectorizer.get_feature_names_out())
print(X_df.iloc[0:5,7555:7565])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_df,df.Spam)
```

```python
y_train=y_train.astype('int')
y_test = y_test.astype('int')
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
print(X_test.shape)
```

```python
pred = model.predict(X_test)
```

```python
print(type(pred))
```

```python
y_test = np.array(y_test).reshape(-1,1)
```

```python
model.score(pred.reshape(-1,1) , y_test)
```
